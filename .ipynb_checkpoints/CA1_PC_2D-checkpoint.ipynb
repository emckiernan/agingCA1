## Defining basic functions

Next, we define various basic functions, including those used to calculate the Boltzmann potential and Nernst potentials for each ion.

```python
# sefine constants
eCharge=1.60217733e-19 # Coulombs
kBoltzmann=1.38065812e-20 # mJ/K
zeroT=273.15 # degree Kelvin

# calculates the Boltzmann or thermal potential at the specified temperature
def vBoltzmann(tempCelcius=37.0):
    return kBoltzmann*(zeroT+tempCelcius)/eCharge

# calculates the Nernst (equilibrium) potential for a given ion
def vNernst(cIn=140.0, cOut=5.0, val=1.0, tempCelcius=37.0):
    vT= kBoltzmann*(zeroT+tempCelcius)/eCharge
    vN= vT * sc.log(cOut/cIn) / val
    return vN

# calculates the Nernst potential normalized by vT
def vTNernst(cIn=140.0, cOut=5.0, val=1.0):
    yN= sc.log(cOut/cIn) / val
    return yN

# calculate the Boltzmann and Nernst potentials using the above functions 
def calcReversalPotentials(parDict):
    parDict['v_T'] = vBoltzmann(parDict['tempCelcius'])
    parDict['v_Na']=vNernst(cIn=parDict['in_Na'], cOut=parDict['out_Na'], val=1.0, \
                            tempCelcius=parDict['tempCelcius'])
    parDict['v_Ka']=vNernst(cIn=parDict['in_Ka'], cOut=parDict['out_Ka'], val=1.0, \
                            tempCelcius=parDict['tempCelcius'])
    parDict['v_ATP']= -450.0;
    parDict['v_NaKa'] = 3*parDict['v_Na'] - 2*parDict['v_Ka'] + parDict['v_ATP']
    return parDict
```

We also define the functions that will be used to calculate ion current flux, ion channel gating, and steady states. 

```python
# transmembrane flux for ionic currents, normalized
def transmFlux_(y,y_r, eta, rect=0.5, A=1.0, degree=sc.e):
    aa= eta*y-y_r
    i=0
    if degree==sc.e:
        i = A * eta *(sc.exp( rect* aa)-sc.exp((rect-1)* aa))
    elif degree==3:
        i = A * eta * aa * (1+ ((rect-1)/2)* aa + ((3* rect **2 -3* rect +1)/6)* (aa**2))
    elif degree==1:
        i =  A * eta * aa
    return i

# used to calculate the rate of activation for potassium channel 
def expSum_(y, s=0.5):
    ee = sc.exp(y)
    return ee**s + ee**(s-1)

# used to calculate the steady states activation cureves for each current 
def expSigmoid_(y):
    ee = sc.exp(y)
    return ee/(1+ee)
```

Next, we define functions to normalize by both volts and current amplitudes, and calculate current amplitudes.

```python
# normalize voltages with respect to the Boltzmann potential kT/q
def normalizeVolts(parDict):
    nDict={}
    vTCm= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vTCm
    for k,i in parDict.items():
        if ((k.find('v_')==0)):
            nDict["y_"+k[2:]]=parDict[k]/parDict['v_T']
    parDict.update(nDict)
    return parDict

# normalizes amplitudes ("a") for transmembrane currents with respect to membrane capacitance 
# new amplitudes are written with capital "A"
def normalizeAmps(parDict):
    nDict={}
    vC= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vC

    for k,i in parDict.items():
        if k.find('a_')==0:
            str0="A_" + k[2:]
            nDict[str0]= i/vC
    parDict.update(nDict)
    return parDict

# calculates amplitudes ("a") for transmembrane currents with respect to membrane capacitance
# new amplitudes are written with capital "A"
def calcAmpsPA(parDict):
    nDict={}
    vC= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vC

    for k,i in parDict.items():
        if k.find('A_')==0:
            str0="a_" + k[2:]
            nDict[str0]= i*vC
    parDict.update(nDict)
    return parDict
```

The following functions are used to create a dictionary with the necessary parameters to run the model, and to print the current amplitudes (normalized and non-normalized).

```python
def addSuffix(p,suff='',removeOriginal=1):
    pa= p.copy()
    for key, val in p.items():
        pa[key+suff]=val
        if removeOriginal:
            pa.pop(key)
    return pa

def prepMembraneDictionary(membraneList,suffixList):
    p= dict()
    p.update(ionsTemperature)
    p.update(numerics)
    nSuff=len(suffixList)
    for n in range(nSuff):
        p.update(addSuffix(membraneList[n],suffixList[n]))
    p= calcReversalPotentials(p)
    p= normalizeAmps(p)
    p['degree']=sc.e
    p= normalizeVolts(p)
    return p

def printAmps(dicti):
    for k,v in dicti.items():
            if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
                print('%s = %g'%(k,v))
    return

def printNormalizedAmps(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
            print('%s = %g'%(k,vv))
    return

def printAmpsPA(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'a_')==0) & (str.find(k,'a_Single_') !=0):
            print('%s = %g'%(k,vv))
    return
```

The following function is used to create a current injection pulse to stimulate the cell. By changing the start and stop time, we can create square or ramp pulses.

```python
# define current injection function
def UpTopDn(t,upStart=200.0,upStop=400, dnStart=600.0,dnStop=800.0,rampAmp=1.0):
    slope_up = rampAmp/(upStop-upStart)
    int_up = -slope_up*upStart
    slope_dn = -rampAmp/(dnStop-dnStart)
    int_dn= rampAmp-slope_dn*dnStart
    c1=sc.int16((upStart<t)&(t<=upStop))
    c2=sc.int16((upStop<t)&(t<=dnStart))
    c3=sc.int16((t>dnStart)&(t<=dnStop))
    y=c1*(slope_up*t +int_up)+ c2*rampAmp+ c3*(slope_dn*t +int_dn)
    return y
```

Next, we define the right-hand side. The model has two variables: the change in voltage (<i>v</i>) and potassium channel activation (<i>w</i>). The currents include a transient sodium current (<i>NaT</i>), a delayed rectifier potassium current (<i>Kd</i>), and a sodium-potassium pump current (<i>NaK</i>).

```python
# define the right-hand side of the model
def NaTKD(U,t,p):    
    x,y=U
    mInf = expSigmoid_(p['gain_Act_NaT']*(y-p['y_Half_Act_NaT']))
    yyKaD = p['gain_Act_KaD']*(y-p['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    xRate = p['rate_Act_KaD'] * expSum_(yyKaD,s=p['bias_Act_KaD'])
    jNa =  transmFlux_(y,-p['y_Na'], eta=p['eta_NaT'], rect=p['rect_NaT'], A=p['A_NaT'], \
                       degree=sc.e)
    jK = transmFlux_(y, p['y_Ka'], eta=p['eta_KaD'], rect=p['rect_KaD'], A=p['A_KaD'], \
                     degree=sc.e)
    jNaKa = transmFlux_(y, p['y_NaKa'], eta=p['eta_NaKa'], rect=p['rect_NaKa'], A=p['A_NaKa'], \
                        degree=sc.e)
    JNa = mInf * (1-x) * jNa
    JK = x * jK
    dy = p['jStim'](t) - JNa - JK - jNaKa
    dx = x * (xInf-x)* xRate
    return sc.array([dx,dy])
```

```python
# solve the system to simulate the electrical activity of the neuron
def simNeuron(p3):
    U= sc.integrate.odeint(func=p3['rhs'], y0=p3['ic'], t=p3['sampTimes'], args=(p3,),\
                           rtol=p3['rTol'],atol=p3['aTol']).transpose()
    oo=dict()
    p3['vTCm']=p3['v_T']*p3['Cm']
    x,y =U 
    v = y * p3['v_T']
    dvdt = sc.zeros(p3['nSteps'])
    dvdt[1:]= (v[1:]-v[:-1])/p3['stepSize']
    oo['x']=x 
    oo['v']=v 
    oo['dv']=dvdt
    mInf = expSigmoid_(p3['gain_Act_NaT']*(y-p3['y_Half_Act_NaT']))
    yyKaD = p3['gain_Act_KaD']*(y-p3['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    p3['xRate'] = p3['rate_Act_KaD'] * expSum_(yyKaD,s=p3['bias_Act_KaD'])
    jNa =  transmFlux_(y,-p3['y_Na'], eta=p3['eta_NaT'], rect=p3['rect_NaT'], \
                       A=p3['A_NaT'], degree=sc.e)
    jK = transmFlux_(y, p3['y_Ka'], eta=p3['eta_KaD'], rect=p3['rect_KaD'], A=p3['A_KaD'], \
                     degree=sc.e)
    jNaKa = transmFlux_(y, p3['y_NaKa'], eta=p3['eta_NaKa'], rect=p3['rect_NaKa'], \
                        A=p3['A_NaKa'], degree=sc.e)
    oo['iNaT'] =(p3['vTCm']*jNa*mInf*(1-x))/1000.0
    oo['iKaD'] =(p3['vTCm']*jK*x)/1000.0
    oo['iNaKa']=(p3['vTCm']*jNaKa)/1000.0
    oo['iStim'] =(p3['vTCm']*p3['jStim'](p3['sampTimes']))
    return oo
```

```python
# set up membrane by creating the dictionary, calculating reversal potentials, normalizing
def membraneSetup(transporters,suffTransp,biophysics):
    p3=dict()
    nTransp = len(transporters); nBiop = len(biophysics)
    for n in range(nTransp):
        p3.update(addSuffix(transporters[n],suff=suffTransp[n],removeOriginal=1))
    for n in range(nBiop):
        p3.update(biophysics[n])
    #p3["A_AK"]= 0.0;
    #p3['y_AK']=0.0; 
    p3 = calcReversalPotentials(p3)
    p3 = normalizeVolts(p3)
    p3 = normalizeAmps(p3)
    p3['vTCm']=p3['v_T']*p3['Cm']
    return p3
```

Next, we create a dictionary with the relevant parameters. The <i>a</i>'s below represent the maximum amplitudes of the currents, which can be thought of as an indicator of the number of functional channels in the membrane. For example, <i>a_NaT</i> represents the sodium channel density. <i>KaD</i> stands for delayed rectifier potassium channel, and <i>NaKa</i> for sodium-potassium pump. The Nernst potentials, as determined by internal and external ion concentrations (e.g. <i>outNa</i>, <i>inNa</i>), remain fixed. Other biophysical channel properties, such as half-activation voltages (e.g. <i>vHalfP_NaT</i>) and gating charges (e.g. <i>gainP_NaT</i>), also remain fixed. The only channel properties that will be varied are the amplitudes (i.e. the number of functional channels in the membrane) and the kinetics (e.g. <i>rateP_KaD</i>).

```python
transp=[biophysNaKa,biophysKaD,biophysNaT]
suffTransp=['_NaKa','_KaD','_NaT','_AK']
biophy=[ionsTemperature,numerics,physicalConstants]
suffBiop = ['','','']
p3 = membraneSetup(transporters=transp,suffTransp=suffTransp,biophysics=biophy)

# dictionary with parameters
def pars(p3):
    p3["Cm"]= 30.0 # membrane capacitance in pF
    # following amplitudes are normalized by membrane capacitance (Cm) and thermopotential (vT)
    # to recover the original amplitudes, multiply by vTCm
    p3['A_NaT']= 1.0 
    p3['A_KaD']= 200.0 
    p3['A_NaKa']= p3['A_NaT']/100.0
    # intracellular and extracellular ion concentrations used to calculate Nernst potentials
    p3['in_Ka']= 140.0
    p3['in_Na']= 10.0
    p3['out_Ka']= 5.0
    p3['out_Na']= 150.0
    # rate of activation of the KD channel
    p3['rate_Act_KaD']= 1.0 
    # symmetry/bias of the time constant of the different ion channels
    p3['bias_Act_KaD']= 0.3
    p3['bias_Act_NaT']= 0.5
    # half-activation voltages for ion channels
    p3['v_Half_Act_KaD']= -5.0 
    p3['v_Half_Act_NaT']= -20.0
    # gating charges for ion channels
    p3['gain_Act_KaD']= 4.0
    p3['gain_Act_NaT']= 4.0
    # time-related parameters and initial conditions
    p3['timeMin']= 0.0
    p3['timeMax'] = 1200.0 
    p3['stepSize']= 0.003     
    p3['nSteps'] = sc.int32(sc.ceil((p3['timeMax']-p3['timeMin'])/p3['stepSize']))
    p3['sampTimes'] = sc.arange(p3['timeMin'],p3["timeMax"],p3["stepSize"])
    p3['ic']=sc.array([0.005,-2.0])
    p3['rhs']=NaTKD
    # stimulus
    p3['upStart']=200.0
    p3['upStop']=200.001
    p3['dnStart']=1000.0
    p3['dnStop']=1000.001
    p3['rampAmp']=100
    # functions
    p3= calcReversalPotentials(p3)
    p3 = normalizeVolts(p3) 
    return p3
```

Now we run the model, first creating the dictionary, then defining the stimulus, and finally solving the system. The normalized and recovered current amplitudes will be printed below, as well as the value of vTCM used for the normalization of the currents.

```python
p3=pars(p3)
printAmps(p3)

p3['jStim'] = lambda tt: (UpTopDn(t=tt,upStart=p3['upStart'],upStop=p3['upStop'],\
                                  dnStart=p3['dnStart'],dnStop=p3['dnStop'],\
                                  rampAmp=p3['rampAmp']))/p3['vTCm']
p3=calcAmpsPA(p3)
printAmpsPA(p3)
oo=simNeuron(p3)
print('The value of vTCm used for normalization of the currents is %g'% (p3['vTCm']))
```

Finally, we graph the result of the above simulation to see the electrical response of the cell to the injected current.

```python
ax1=list()
fig1=plt.figure(figsize=(15,11))
plt.ioff()
r=2; c=1
for s in sc.arange(r*c):
    ax1.append(fig1.add_subplot(r,c,s+1))

ax1[0].plot(p3['sampTimes'],oo['v'],'k')
ax1[0].set_ylabel('V (mV)')
ax1[0].set_ylim(-80,40)
ax1[1].plot(p3['sampTimes'],oo['iStim'],'k',label=r'$I_{Stim}$')
ax1[1].set_ylabel('$I_{Stim}$ (pA)')
ax1[1].set_ylim(-3,200)
ax1[1].set_xlabel('time (ms)');
```
