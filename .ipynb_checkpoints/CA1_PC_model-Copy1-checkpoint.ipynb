# Overview

We are interested in modeling age-related changes in ion channel expression and excitability in CA1 pyramidal cells (PCs). However, before we can model aging, we have to create a representative young CA1 PC. PCs in the CA1 region of the hippocampus show a diversity of firing patterns, including repetitive slow firing with frequency adaptation, conditional bursting (bursting in response only to a stimulus), and endogenous bursting in the absence of stimulation. We can build a 3-dimensional model, based on electrophysiological data, to determine under what conditions these different firing patterns can be reproduced. We will vary primarily the density of ion channels in the model membrane and specific biophysical properties of the channels, such as their rate of activation. In this way, we can understand the different balances of currents that produce these firing patterns and then compare the firing in young and aged PCs. We assume that the passive membrane properties of young and aged cells are identical and that these cells primarily differ in their calcium channel density, as shown in the experimental literature.

## Defining basic functions

Now we define various basic functions, including those used to calculate the Boltzmann potential and Nernst potentials for each ion.

```python
# define constants
eCharge=1.60217733e-19 # Coulombs
kBoltzmann=1.38065812e-20 # mJ/K
zeroT=273.15 # degree Kelvin

# calculates the Boltzmann or thermal potential at the specified temperature
def vBoltzmann(tempCelcius=37.0):
    return kBoltzmann*(zeroT+tempCelcius)/eCharge

# calculates the Nernst (equilibrium) potential for a given ion
def vNernst(cIn=140.0, cOut=5.0, val=1.0, tempCelcius=37.0):
    vT= kBoltzmann*(zeroT+tempCelcius)/eCharge
    vN= vT * sc.log(cOut/cIn) / val
    return vN

# calculates the Nernst potential normalized by vT
def vTNernst(cIn=140.0, cOut=5.0, val=1.0):
    yN= sc.log(cOut/cIn) / val
    return yN

# calculate the Boltzmann and Nernst potentials using the above functions 
def calcReversalPotentials(parDict):
    parDict['v_T'] = vBoltzmann(parDict['tempCelcius'])
    parDict['v_Ca']=vNernst(cIn=parDict['in_Ca'], cOut=parDict['out_Ca'], val=2.0, \
                            tempCelcius=parDict['tempCelcius'])
    parDict['v_Na']=vNernst(cIn=parDict['in_Na'], cOut=parDict['out_Na'], val=1.0, \
                            tempCelcius=parDict['tempCelcius'])
    parDict['v_Ka']=vNernst(cIn=parDict['in_Ka'], cOut=parDict['out_Ka'], val=1.0, \
                            tempCelcius=parDict['tempCelcius'])
    parDict['v_ATP']= -450.0;
    parDict['v_NaKa'] = 3*parDict['v_Na'] - 2*parDict['v_Ka'] + parDict['v_ATP']
    return parDict
```

We also define the functions that will be used to calculate ion current flux, ion channel gating, and steady states.

```python
# transmembrane ion flux, normalized 
def transmFlux_(y,y_r, eta, rect=0.5, A=1.0, degree=sc.e):
    aa= eta*y-y_r
    i=0
    if degree==sc.e:
        i = A * eta *(sc.exp( rect* aa)-sc.exp((rect-1)* aa))
    elif degree==3:
        i = A * eta * aa * (1+ ((rect-1)/2)* aa + ((3* rect **2 -3* rect +1)/6)* (aa**2))
    elif degree==1:
        i =  A * eta * aa
    return i

def expSum_(y, s=0.5):
    ee = sc.exp(y)
    return ee**s + ee**(s-1)

def expSigmoid_(y):
    ee = sc.exp(y)
    return ee/(1+ee)

def wInverse_(w,vHalf,gCharge,vT):
    v= vHalf - (vT/gCharge) * sc.log( (1-w)/w)
    return v

def expSub_(y, s=0.5):
    ee = sc.exp(y)
    return ee**s - ee**(s-1)

def logisticDynamics_(t,u0,a,r):
    return (a*u0)/(u0 + (a-u0)*sc.exp(-a*r*t))
```

Next, we define functions to normalize by both volts and current amplitudes, and calculate current amplitudes.

```python
# normalize voltages with respect to the Boltzmann potential kT/q
def normalizeVolts(parDict):
    nDict={}
    vTCm= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vTCm
    for k,i in parDict.items():
        if ((k.find('v_')==0)):
            #print(k)
            nDict["y_"+k[2:]]=parDict[k]/parDict['v_T']
    parDict.update(nDict)
    return parDict

# normalizes amplitudes ("a") for transmembrane currents with respect to the membrane capacitance 
# new amplitudes are written with capital "A"
def normalizeAmps(parDict):
    nDict={}
    vC= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vC

    for k,i in parDict.items():
        if k.find('a_')==0:
            #print(k,i)
            str0="A_" + k[2:]
            #if type(i)==float:
            nDict[str0]= i/vC
            #print(nDict[str0])
    parDict.update(nDict)
    return parDict

# calculates amplitudes ("a") for transmembrane currents with respect to the membrane capacitance
# new amplitudes are written with capital "A"
def calcAmpsPA(parDict):
    nDict={}
    vC= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vC

    for k,i in parDict.items():
        if k.find('A_')==0:
            #print(k,i)
            str0="a_" + k[2:]
            #if type(i)==float:
            nDict[str0]= i*vC
            #print(nDict[str0])
    parDict.update(nDict)
    return parDict

# normalizes amplitudes ("a") for transmembrane currents with respect to the membrane capacitance
# new amplitudes are written with capital "A"
def normalizeSingleAmps(parDict):
    nDict={}
    vC= parDict['Cm'] * parDict['v_T']
    nDict['vTCm']=vC

    for k,i in parDict.items():
        if k.find('a_Single')==0:
            #print(k,i)
            str0="A_" + k[2:]
            #if type(i)==float:
            nDict[str0]= i/vC
            #print(nDict[str0])
    parDict.update(nDict)
    return parDict
```

The following functions are used to create a dictionary with the necessary parameters to run the model, and to print the current amplitudes (normalized and non-normalized) in particular.

```python
def addSuffix(p,suff='',removeOriginal=1):
    pa= p.copy()
    for key, val in p.items():
        pa[key+suff]=val
        if removeOriginal:
            pa.pop(key)
    return pa

def prepMembraneDictionary(membraneList,suffixList):
    p= dict()
    p.update(ionsTemperature)
    p.update(numerics)
    nSuff=len(suffixList)
    for n in range(nSuff):
        p.update(addSuffix(membraneList[n],suffixList[n]))
    p= calcReversalPotentials(p)
    p= normalizeAmps(p)
    p['degree']=sc.e
    p= normalizeVolts(p)
    return p

def printAmps(dicti):
    for k,v in dicti.items():
            if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
                print('%s = %g'%(k,v))
    return

def printNormalizedAmps(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
            print('%s = %g'%(k,vv))
    return

def printAmpsPA(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'a_')==0) & (str.find(k,'a_Single_') !=0):
            print('%s = %g'%(k,vv))
    return
```

If we do not want to use the built-in solver provided by scipy, we can define a custom second-order Runge-Kutta solver (as below). Currently we use odeint due to speed of simulations.

```python
# second-order Runge-Kutta method to solve x' = f(x) with U(t[0]) = U0
# [ECM comment: this solver runs slower than built-in solver odeint; currently using odeint] 
def RK2_Autonomous(f, p, parNames=[],parValues=[]):
    nForc=len(parNames)
    U=sc.zeros((p['nSteps'], sc.prod(sc.shape(p['ic']))),"float64")
    U[0]=p['ic']
    if nForc>0:
        for i in sc.arange(p['nSteps']-1):
            for nn in sc.arange(nForc):
                p[parNames[nn]]=parValues[nn][i]
            k1 = p['stepSize'] * f( U[i], p) / 2.0
            U[i+1] = U[i] + p['stepSize'] * f( U[i] + k1, p)
    else:
        for i in sc.arange(p['nSteps']-1):
            k1 = p['stepSize'] * f( U[i], p) / 2.0
            U[i+1] = U[i] + p['stepSize'] * f( U[i] + k1, p)
    return U.transpose()
```

Now we define the current injection function used to stimulate model neurons as in experiments. The timing of the pulse can be modified to simulate a square pulse, or ramps of different slopes.

```python
# define current injection function
def UpTopDn(t,upStart=200.0,upStop=400, dnStart=600.0,dnStop=800.0,rampAmp=1.0):
    slope_up = rampAmp/(upStop-upStart)
    int_up = -slope_up*upStart
    slope_dn = -rampAmp/(dnStop-dnStart)
    int_dn= rampAmp-slope_dn*dnStart
    c1=sc.int16((upStart<t)&(t<=upStop))
    c2=sc.int16((upStop<t)&(t<=dnStart))
    c3=sc.int16((t>dnStart)&(t<=dnStop))
    y=c1*(slope_up*t +int_up)+ c2*rampAmp+ c3*(slope_dn*t +int_dn)
    return y
```

Next, we define the right-hand side. The model has three variables: the change in voltage (<i>v</i>), potassium channel activation (<i>w</i>), and intracellular calcium concentration (<i>c</i>). The currents include a transient sodium current (<i>NaT</i>), a delayed rectifier potassium current (<i>KaD</i>), a L-type calcium current (<i>CaL</i>), a calcium-dependent potassium current (<i>KaSK</i>), and a sodium-potassium pump (<i>NaKa</i>).

```python
# define the right-hand side of the model
def NaTKDSKCaL(U,t,p):    
    x,y,c=U
    p['y_Ca']= vTNernst_(cIn=c, cOut=p['out_Ca'], val=2.0)
    yyKaD = p['gain_Act_KaD']*(y-p['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    xRate = p['rate_Act_KaD'] * expSum_(yyKaD,s=p['bias_Act_KaD'])
    m13Inf = expSigmoid_(p['gain_Act_CaL']*(y-p['y_Half_Act_CaL']))
    mInf = expSigmoid_(p['gain_Act_NaT']*(y-p['y_Half_Act_NaT'])) 
    cc= c**p['gain_Act_KaSK']
    cDenom= cc + p['c_Half_Act_KaSK']**p['gain_Act_KaSK']
    skInf= cc/cDenom
    jNaKa = transmFlux_(y, p['y_NaKa'], eta=p['eta_NaKa'], rect=p['rect_NaKa'], A=p['A_NaKa'], \
                        degree=sc.e)
    jNa =  transmFlux_(y,-p['y_Na'], eta=p['eta_NaT'], rect=p['rect_NaT'], A=p['A_NaT'], \
                       degree=sc.e)
    jCaL = transmFlux_(y, -p['y_Ca'], eta=p['eta_CaL'], rect=p['rect_CaL'], A=p['A_CaL'], \
                       degree=sc.e)
    jK = transmFlux_(y, p['y_Ka'], eta=p['eta_KaD'], rect=p['rect_KaD'], A=1.0, \
                     degree=sc.e)
    """Patch recordings indicate that the L-type channels in these cells have little to no inactivation. Commenting
    out and putting in a non-activating version of the current instead. Running the comparison with and without
    inactivating, only very subtle differences in firing frequency observed."""
    # JCa = (1-x) * m13Inf * jCaL ; 
    JCa = m13Inf * jCaL
    JNa = mInf * (1-x) * jNa
    JK = (p['A_KaSK'] * skInf + p['A_KaD'] * x ) * jK
    #dy = p['A_F'] - JNa - JCa - JK - jNaKa
    dy = p['jStim'](t) - JNa - JCa - JK - jNaKa
    dx = x * (xInf-x)* xRate
    dc= p['r_Ca'] * (p['c_Min_Ca']-c) - p['k_Ca']* JCa
    return sc.array([dx,dy,dc])
```

Now we define the function which will solve the system of equations and simulate electrical activity in the model neuron.

```python
#def simNeuron(p3,pNames,pValues):
def simNeuron(p3):
    #U = RK2_Autonomous(f=NaTKDSKCaL, p=p3, parNames=pNames,parValues=pValues)
    U= sc.integrate.odeint(func=p3['rhs'], y0=p3['ic'], t=p3['sampTimes'], args=(p3,),rtol=p3['rTol'],\
                           atol=p3['aTol']).transpose()
    oo=dict()
    p3['vTCm']=p3['v_T']*p3['Cm']
    x,y,c =U; v = y * p3['v_T'];
    dvdt = sc.zeros(p3['nSteps'])
    dvdt[1:]= (v[1:]-v[:-1])/p3['stepSize']
    oo['x']=x; oo['v']=v; oo['dv']=dvdt; oo['c']=c
    oo['y_Ca']= vTNernst_(cIn=c, cOut=p3['out_Ca'], val=2.0)
    oo['v_Ca']= oo['y_Ca']*p3['v_T']
    yyKaD = p3['gain_Act_KaD']*(y-p3['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    p3['xRate'] = p3['rate_Act_KaD'] * expSum_(yyKaD,s=p3['bias_Act_KaD'])
    m13Inf = expSigmoid_(p3['gain_Act_CaL']*(y-p3['y_Half_Act_CaL']))
    mInf = expSigmoid_(p3['gain_Act_NaT']*(y-p3['y_Half_Act_NaT']))
    cc= c**p3['gain_Act_KaSK']
    cDenom= cc + p3['c_Half_Act_KaSK']**p3['gain_Act_KaSK']
    skInf= cc/ cDenom
    jNaKa = transmFlux_(y, p3['y_NaKa'], eta=p3['eta_NaKa'], rect=p3['rect_NaKa'], \
                        A=p3['A_NaKa'], degree=sc.e)
    jNa =  transmFlux_(y,-p3['y_Na'], eta=p3['eta_NaT'], rect=p3['rect_NaT'], \
                       A=p3['A_NaT'], degree=sc.e)
    jCaL = transmFlux_(y, -oo['y_Ca'], eta=p3['eta_CaL'], rect=p3['rect_CaL'], \
                       A=p3['A_CaL'], degree=sc.e)
    jK = transmFlux_(y, p3['y_Ka'], eta=p3['eta_KaD'], rect=p3['rect_KaD'], A=1.0, \
                     degree=sc.e)
    """Have changed the modeling for the L-type current, removing inactivation, as per comments above."""
    #dy = p3['A_F']- mInf * (1-x) * jNa - (1-x) * m13Inf * jCaL - (p3['A_KaSK'] * skInf \
                                                                    # + p3['A_KaD'] * x ) * jK - jNaKa
    #oo['iTot'] = -p3['v_T']*p3['Cm']* dy
    oo['iNaKa']=(p3['vTCm']*jNaKa)/1000.0
    oo['iKaD'] =(p3['vTCm']*jK*p3['A_KaD']*x)/1000.0
    oo['iKaSK']=(p3['vTCm']*jK*p3['A_KaSK'] * skInf)/1000.0
    oo['iNaT'] =(p3['vTCm']*jNa*mInf * (1-x))/1000.0
    oo['iCaL'] =(p3['vTCm']*jCaL*m13Inf * (1-x))/1000.0
    oo['iStim'] = (p3['jStim'](p3['sampTimes']))*p3['vTCm']
    return oo
```

Finally, we define and run functions used to create a dictionary with all the necessary parameters. We first setup the membrane using some default parameters from our Membrane Library. 

```python
def membraneSetup(transporters,suffTransp,biophysics):
    p3=dict()
    nTransp = len(transporters); nBiop = len(biophysics)
    for n in range(nTransp):
        p3.update(addSuffix(transporters[n],suff=suffTransp[n],removeOriginal=1))
    for n in range(nBiop):
        p3.update(biophysics[n])
    #
    p3["A_F"]= 0.0;
    p3["A_AK"]= 0.0;
    p3["A_GA"]= 0.0;
    p3['y_AK']=0.0; 
    p3['y_GA']=0.0;
    p3 = calcReversalPotentials(p3)
    p3 = normalizeVolts(p3)
    p3 = normalizeAmps(p3)
    p3['vTCm']=p3['v_T']*p3['Cm']
    return p3

transp=[biophysNaKa,biophysKaD,biophysNaT,biophysKaSK,biophysCa13,biophysCalcium,biophysAMPA,biophysGabaA]
suffTransp=['_NaKa','_KaD','_NaT','_KaSK','_CaL','_Ca','_AK','_GA']
biophy=[ionsTemperature,numerics,physicalConstants]
suffBiop = ['','','']
p3 = membraneSetup(transporters=transp,suffTransp=suffTransp,biophysics=biophy)
```

Then we redefine below the bulk of the relevant parameters. This is done explicitly here so we can see what ion and channel settings are and then change these as needed to create different firing patterns.

```python
# setup dictionary with parameters
def pars(p3):
    p3["Cm"]= 25.0 # membrane capacitance in pF
    # following amplitudes are normalized by the membrane capacitance (Cm) and thermopotential (vT)
    # to recover the original amplitudes, multiply by vTCm
    p3['A_NaT']= 1.5 
    p3['A_KaD']= 40.0
    p3['A_KaSK']= 1.1
    p3['A_CaL']= 0.4
    p3['A_NaKa']= p3['A_NaT']/100.0
    # intracellular and extracellular ion concentrations used to calculate Nernst potentials
    p3['in_Ca']= 0.0001
    p3['in_Ka']= 140.0
    p3['in_Na']= 13.0 #10.0
    p3['out_Ca']= 2.5
    p3['out_Ka']= 5.0
    p3['out_Na']= 150.0
    p3['rate_Act_KaD']= 1.0; # rate of activation of the KD channel
    # symmetry/bias of the different ion channels
    p3['bias_Act_CaL']= 0.5
    p3['bias_Act_KaD']= 0.3
    p3['bias_Act_KaSK']= 0.3
    p3['bias_Act_NaT']= 0.5
    # half-activation voltages (or concentration for SK) for the different ion channels
    p3['v_Half_Act_KaD']= -1.0 
    p3['v_Half_Act_NaT']= -19.0
    p3['v_Half_Act_CaL']= 3.0
    p3['c_Half_Act_KaSK']= 7.4e-4 # micromolar 
    # gating charges for the different ion channels
    p3['gain_Act_KaD']= 4.0
    p3['gain_Act_NaT']= 4.0
    p3['gain_Act_CaL']= 4.0
    p3['gain_Act_KaSK']= 2.0
    # calcium buffering, intracellular handling 
    p3['r_Ca']= 1e-3 
    p3['k_Ca']= 8e-6 
    p3['c_Min_Ca']= 1e-4 # micromolar
    # time-related parameters and initial conditions
    p3['timeMin']= 0.0
    p3['timeMax'] = 1200.0 
    p3['stepSize']= 0.002     
    p3['nSteps'] = sc.int32(sc.ceil((p3['timeMax']-p3['timeMin'])/p3['stepSize']))
    p3['sampTimes'] = sc.arange(p3['timeMin'],p3["timeMax"],p3["stepSize"])
    p3['ic']=sc.array([0.005,-2.8,1e-4])
    p3['rhs']=NaTKDSKCaL
    # stimulus parameters; start/stop times and amplitude
    p3['upStart']=200.0
    p3['upStop']=200.001
    p3['dnStart']=1000.0
    p3['dnStop']=1000.001
    p3['rampAmp']=100
    # functions to calculate Nernst potentials and normalize
    p3= calcReversalPotentials(p3)
    p3 = normalizeVolts(p3) 
    return p3
```

# Simulating CA1 PC electrical activity

## Repetitive slow firing with spike frequency adaptation

Many CA1 pyramidal cells respond to square-pulse current injection by firing several early spikes followed by marked adaptation, which slows the frequency of firing. In aged animals, this adaptation is more pronounced than in young animals, leading to a shorter initial period of fast spiking, followed by fewer spikes or even complete cessation of activity. The increased adaptation seen in aged PCs is functionally relevant, as it is correlated with learning impairment. Therefore, our first challenge was to tune the model to produce early-onset firing with spike frequency adaptation. The baseline parameter regime specified above produces this firing pattern (see full paper for more details). The following commands run the model:

```python
p3=pars(p3)
printAmps(p3)
#AF=sc.zeros(len(p3['sampTimes']))
#a=sc.int32(200./p3['stepSize']) 
#b=sc.int32(1000./p3['stepSize'])
#AF[a:b]=0.19
#p3['A_F']=AF.max() 
#p3['a_F_pA/pF']=p3['A_F']*p3['v_T'] 
#p3['a_F_pA']=p3['A_F']*p3['vTCm']
p3['jStim'] = lambda tt: (UpTopDn(t=tt,upStart=p3['upStart'],upStop=p3['upStop'], dnStart=p3['dnStart'],\
                                  dnStop=p3['dnStop'],rampAmp=p3['rampAmp']))/p3['vTCm']
p3=calcAmpsPA(p3)
printAmpsPA(p3)
#oo=simNeuron(p3,['A_F'],[AF])
oo=simNeuron(p3)
#print('The stimulus amplitude is %g pA (or %g pA/pF, or %g without units)'% (p3['a_F_pA'],p3['a_F_pA/pF'],p3['A_F']))
print('The value of vTCm used for normalization of the currents is %g'% (p3['vTCm']))
```

## Afterhyperpolarizations

CA1 PCs stimulated with a short square-pulse current injection into the soma can be induced to fire a quick burst of spikes, followed by a period during which the membrane potential dips below the resting membrane potential and remains there for hundres of milliseconds to seconds. This period is known as an after-hyperpolarization (AHP). Experimental studies have shown that aged CA1 PCs have larger AHPs than young cells, causing decreased excitability and spike failure. 

For the following simulations, the amplitudes of the Na<sup>+</sup>, delayed rectifier K<sup>+</sup>, and Ca<sup>2+</sup>-dependent K<sup>+</sup> currents were all increased relative to the previous simulations, but still within the physiological range. (Note: Don't forget to also reset the Na<sup>+</sup>/K<sup>+</sup> pump current amplitude, since it is a multiple of <i>A_NaT</i>.)  The amplitude of the Ca<sup>2+</sup> current and Ca<sup>2+</sup> handling was not changed. We then stimulated the model cells with a 100ms square pulse current injection such that it induced a burst of four spikes in both young and aged cells. This required the current amplitude to be increased (40 pA higher) for aged cells. Since the same number of spikes is generated in each cell, their subsequent AHPs can be compared. To compare young and aged cells, we once again fixed all parameters except for the amplitude of the L-type Ca<sup>2+</sup> current, which was set to produce currents of around 3 nA or 5 nA, respectively. 

```python
#set parameters for AHPs
p3=pars(p3)
p3['A_NaT']= 2.0 
p3['A_KaD']= 50.0 
p3['A_KaSK']= 1.3 
p3['A_CaL']= 0.4
p3['A_NaKa']= p3['A_NaT']/100.0
p3['timeMin']= 0.0
p3['timeMax'] = 8000.0 
p3['stepSize']= 0.003     
p3['nSteps'] = sc.int32(sc.ceil((p3['timeMax']-p3['timeMin'])/p3['stepSize']))
p3['sampTimes'] = sc.arange(p3['timeMin'],p3["timeMax"],p3["stepSize"])
p3['upStart']=200.0
p3['upStop']=200.001
p3['dnStart']=300.0
p3['dnStop']=300.001

ACaLs=sc.arange(0.4,0.8,0.3)
stims=sc.arange(80,130,40)
nstims=len(stims)
sims=list()

for n in sc.arange(nstims):
    p3['A_CaL']=ACaLs[n]
    p3['rampAmp']=stims[n]
    #AF=sc.zeros(len(p3['sampTimes']))
    #a=sc.int32(200./p3['stepSize'])
    #b=sc.int32(300./p3['stepSize']);
    #AF[a:b]=stims[n] 
    #p3['A_F']=AF.max(); p3['a_F_pA/pF']=p3['A_F']*p3['v_T']; p3['a_F_pA']=p3['A_F']*p3['vTCm']
    p3['jStim'] = lambda tt: (UpTopDn(t=tt,upStart=p3['upStart'],upStop=p3['upStop'], dnStart=p3['dnStart'],\
                                  dnStop=p3['dnStop'],rampAmp=p3['rampAmp']))/p3['vTCm']
    p3=calcAmpsPA(p3); printAmpsPA(p3)
    #oo=simNeuron(p3,['A_F'],[AF])
    oo=simNeuron(p3)
    sims.append(oo) 
```

## Local field potential forcing

While square pulse stimulation is useful for examining the timing of neural responses, it is not a physiologically realistic stimulus. Instead, we can define an Ornstein-Uhlenbeck stochastic process to simulate local field potential (LFP) forcing that CA1 PCs could experience due to the electrical activity of cells in their vicinity.

```python
# define OU forcing function
def ouForcing(pa,suffix='_LFP',graph=0):
    if pa['aSD'+suffix]>0:
        diffF=2*pa['aSD'+suffix]/pa['aTau'+suffix]
        ouProcess=sp.OU_process(theta=1/pa['aTau'+suffix], mu=pa['aMean'+suffix], sigma=diffF)
        st= sc.copy(pa['sampTimes'])
        ouPath= ouProcess.sample_path(st)[0]
        ouFunc=lambda t: sc.interp(t,st,ouPath)
        pa['aOU'+suffix]=ouProcess
        pa['aOUFunc'+suffix]=ouFunc
        pa['aOUPath'+suffix]=ouPath
        if graph:
            plt.figure()
            plt.plot(st, ouPath, label=r'$a_{%s}$'%suffix)
            plt.legend()
    else:
        print('Could not simulate OU process')
    return pa
```

Now, we redefine the right-hand side to include the LFP forcing.

```python
# define right-hand side to include LFP (OU) forcing
def NaTKDSKCaLOU(U,t,p):  
    x,y,c=U
    p['y_Ca']= vTNernst_(cIn=c, cOut=p['out_Ca'], val=2.0)
    yyKaD = p['gain_Act_KaD']*(y-p['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    xRate = p['rate_Act_KaD'] * expSum_(yyKaD,s=p['bias_Act_KaD'])
    m13Inf = expSigmoid_(p['gain_Act_CaL']*(y-p['y_Half_Act_CaL']))
    mInf = expSigmoid_(p['gain_Act_NaT']*(y-p['y_Half_Act_NaT'])) 
    cc= c**p['gain_Act_KaSK']
    cDenom= cc + p['c_Half_Act_KaSK']**p['gain_Act_KaSK']
    skInf= cc/cDenom
    jNaKa = transmFlux_(y, p['y_NaKa'], eta=p['eta_NaKa'], rect=p['rect_NaKa'], A=p['A_NaKa'], \
                        degree=sc.e)
    jNa =  transmFlux_(y,-p['y_Na'], eta=p['eta_NaT'], rect=p['rect_NaT'], A=p['A_NaT'], \
                       degree=sc.e)
    jCaL = transmFlux_(y, -p['y_Ca'], eta=p['eta_CaL'], rect=p['rect_CaL'], A=p['A_CaL'], \
                       degree=sc.e)
    jK = transmFlux_(y, p['y_Ka'], eta=p['eta_KaD'], rect=p['rect_KaD'], A=1.0, \
                     degree=sc.e)
    JCa = m13Inf * jCaL
    JNa = mInf * (1-x) * jNa
    JK = (p['A_KaSK'] * skInf + p['A_KaD'] * x ) * jK
    jLFP = p['aOUFunc_LFP'](t)/p['vTCm']
    dy = p['jStim'](t) - JNa - JCa - JK - jNaKa - jLFP
    dx = x * (xInf-x)* xRate
    dc= p['r_Ca'] * (p['c_Min_Ca']-c) - p['k_Ca']* JCa
    return sc.array([dx,dy,dc])
```
