```python
# -----------------------
# Evolution rules:
# The variables (w,v) are changed to (x,y) with y=v/v_T and x as in w but depending on y instead of v
# -----------------------
def PyrCA1_LFP(U,p):
    x,y,c=U
    p['y_Ca']= np.log(p['out_Ca']/c)/2.0
    yyKaD = p['gain_Act_KaD']*(y-p['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    xRate = p['rate_Act_KaD'] * expSum_(yyKaD,s=p['bias_Act_KaD'])
    pCaL = expSigmoid_(p['gain_Act_CaL']*(y-p['y_Half_Act_CaL']))
    pNaT = (1-x)*expSigmoid_(p['gain_Act_NaT']*(y-p['y_Half_Act_NaT']))
    pSK= MM(c,p['c_Half_Act_KaSK'],p['gain_Act_KaSK'])
    AK = p['A_KaSK'] * pSK + p['A_KaD'] * x
    jKa = AK * np.sinh((y-p['y_Ka'])/2.)
    jNaKa = p['A_NaKa'] * np.sinh((y-p['y_NaKa'])/2.)
    jNaT = p['A_NaT'] * pNaT * np.sinh((y-p['y_Na'])/2.)
    jCaL = p['A_CaL'] * pCaL * np.sinh( y - p['y_Ca'])
    dx = x * (xInf-x)* xRate ;
    dy = p['A_F'] - jNaT - jCaL - jKa - jNaKa
    dc = (p['c_Min_Ca']-c)/p['tau_Ca'] - (p['k_Ca'] * jCaL) 
    return np.array([dx,dy,dc])
```

```python
# -----------------------
# Numerical solution and calculations
# -----------------------
def numericalSolution_PyrCA1(rhs,p,parNames,parValues):
    start = time()
    U =autonomousRK2_Solver(f=rhs, p=p, parNames=parNames,parValues=parValues)
    oo={'solution':U,'timeSamples':p['timeSamples']}
    x,y,c =U; v = y * p['v_T'];
    dvdt = np.zeros(len(v))
    dvdt[1:]= (v[1:]-v[:-1])/p['stepSize']
    print('Max \partial_t v = %g'%dvdt.max())
    oo['y']=y
    oo['x']=x; oo['v']=v; oo['\partial_t v']=dvdt; oo['c']=c
    oo['v_Ca']= vNernst_(cOut=p['out_Ca'], cIn=c, val=2.0, tempCelcius=p3['tempCelcius'])
    oo['y_Ca'] = np.log(p['out_Ca']/c)/2.0
    yyKaD = p['gain_Act_KaD']*(y-p['y_Half_Act_KaD'])
    xInf = expSigmoid_(yyKaD)
    xRate = p['rate_Act_KaD'] * expSum_(yyKaD,s=p['bias_Act_KaD'])
    pCaL = expSigmoid_(p['gain_Act_CaL']*(y-p['y_Half_Act_CaL']))
    pNaT = (1-x)*expSigmoid_(p['gain_Act_NaT']*(y-p['y_Half_Act_NaT']))
    pKaSK= MM(c,p['c_Half_Act_KaSK'],p['gain_Act_KaSK'])
    iNaKa = p['vTCm']*p['A_NaKa'] * np.sinh((y-p['y_NaKa'])/2)
    iNaT = p['vTCm']*p['A_NaT'] * pNaT * np.sinh((y-p['y_Na'])/2)
    iKaD = p['vTCm']* p['A_KaD'] * x * np.sinh((y-p['y_Ka'])/2)
    iKaSK = p['vTCm']*p['A_KaSK'] * pKaSK * np.sinh((y-p['y_Ka'])/2)
    iCaL = p['vTCm']*p['A_CaL'] * pCaL * np.sinh(y-p['y_Ca'])
    #
    print('(Promedio) vCa=%g mV'%(p['v_T'] * np.log(p['out_Ca']/c)/2.0).mean())
    oo['p_KaSK'] = pKaSK
    oo['p_NaT'] = pNaT
    oo['p_CaL'] = pCaL
    oo['i_NaKa']= iNaKa
    oo['i_KaD'] = iKaD
    oo['i_KaSK']= iKaSK
    oo['i_NaT'] = iNaT
    oo['i_CaL'] = iCaL
    #oo['iStim'] = (p3['jStim'](p3['timeSamples']))*p3['vTCm']
    print('%d seconds'% (time()-start))
    return oo
```

```python
def membraneSetup(transporters,suffTransp,biophysics):
    p3=dict()
    nTransp = len(transporters); nBiop = len(biophysics)
    for n in range(nTransp):
        p3.update(addSuffix(transporters[n],suff=suffTransp[n],removeOriginal=1))
    for n in range(nBiop):
        p3.update(biophysics[n])
    #
    p3["A_F"]= 0.0;
    p3["A_AK"]= 0.0;
    p3["A_GA"]= 0.0;
    p3 = calcReversalPotentials(p3)
    p3 = normalizeVolts(p3)
    p3['y_AK']=0.0/p3['v_T'];
    p3['y_GA']=-70.0/p3['v_T'];
    p3 = normalizeAmps(p3)
    p3['vTCm']=p3['v_T']*p3['Cm']
    return p3
```

```python
# setup dictionary with parameters
def pars_PyrCA1(p3, profile='condBursting',age='young'):
    p3['degree'] = 2.8182
    p3['tempCelcius']=37.0
    p3["Cm"]= 25.0 # membrane capacitance in pF
    p3['v_T']=kBoltzmann*(zeroT+p3['tempCelcius'])/eCharge
    p3['vTCm']= p3['Cm'] * p3['v_T']
    # amplitudes normalized by v_T \cdot C_m
    # to recover the original amplitudes, multiply by vTCm
    p3['A_NaKa']= p3['A_NaT']/100.0
    p3['A_AK']= 0.0
    p3['A_GA']= 0.0
    p3['A_F']= 0.0
    # intracellular and extracellular ion concentrations used to calculate Nernst potentials
    p3['in_Ca']= 1e-4 #uM equiv 0.1 nM
    p3['in_Ka']= 140.0 #mM
    p3['in_Na']= 13.0 #mM
    p3['out_Ca']= 1.5e3 #uM equivalent to 1.5 mM
    p3['out_Ka']= 5.0 #mM
    p3['out_Na']= 150.0 #mM
    # rate of activation of the KD channel
    #p3['rate_Act_KaD']= 2.5;
    # symmetry/bias of the different ion channels
    p3['bias_Act_CaL']= 0.5
    p3['bias_Act_KaD']= 0.4
    p3['bias_Act_KaSK']= 0.5
    p3['bias_Act_NaT']= 0.5
    # half-activation voltages (or concentration for SK) for the different ion channels
    p3['v_Half_Act_KaD']= -1.0
    p3['v_Half_Act_NaT']= -20.0
    p3['v_Half_Act_CaL']= 3.0
    p3['c_Half_Act_KaSK']= 0.3 #0.0003 mM = 0.3 uM Stocker, 2004, Ca-activated K SK channels
    # gating charges for the different ion channels
    p3['gain_Act_KaD']= 4.0
    p3['gain_Act_CaL']= 4.0
    p3['gain_Act_NaT']= 6.0
    p3['gain_Act_KaSK']= 3
    # activation/deactivation of postsynaptic currents
    p3['alpha_AK'] = 0.0
    p3['alpha_GA'] = 0.0
    p3['beta_AK'] = 0.19
    p3['beta_GA'] = 0.18
    # calcium buffering, intracellular handling
    p3['tau_Ca']= 75.0
    p3['rate_Ca']= 1/p3['tau_Ca']
    #p3['k_Ca']= 6e-6 # needs adjusting with respect to the normalized amplitude of the Ca current
    p3['c_Min_Ca']= 1e-4 #uM, 1e-4 means 0.1 nmol
    # Profiles
    if profile=='spontBursting':
        p3['A_NaT']= 2.0; p3['A_NaK']= p3['A_NaT']/100.0      
        p3['gain_Act_NaT']= 4.0; p3['y_Half_Act_NaT']= -21.0/p3['v_T']
        p3['A_KaD'] = 13.0;
        p3['gain_Act_KaD']=4.0; p3['bias_Act_KaD'] = 0.35; p3['rate_Act_KaD']=0.7
        p3['A_KaSK']= 2; p3['gain_Act_KaSK']= 3; p3['c_Half_Act_KaSK'] = 0.3
        #
        p3['tau_Ca']=75.0; p3['k_Ca']=1e-2;
        if age=='young':
            p3['A_CaL']=0.02; 
        elif age=='old':
            p3['A_CaL']=0.04; 

    # Adaptive, conditional firing profile
    # Rheobase at approx 150 pA
    # Resembling Fig. 1 from Disterhoft (1996)
    elif profile=='CA1_adaptiveFiring1':
        p3['A_NaT']= 1.5
        AKTot = 10; wKaD= 9/11
        p3['A_KaD']= AKTot*wKaD; p3['rate_Act_KaD']= 0.9;
        p3['A_KaSK']= AKTot*(1-wKaD);
        p3['A_NaKa'] = p3['A_NaT']/100.0
        p3['tau_Ca']=75.0; p3['k_Ca']=0.7e-4;

    # Adaptive, conditional firing profile
    # Rheobase at approx 150 pA
    elif profile=='adaptiveFiring2':
        p3['A_NaT']= 2.0
        p3['A_KaD']= 12; p3['rate_Act_KaD']= 0.9;
        p3['A_KaSK']= 0.4
        p3['A_CaL']= 0.01
        p3['tau_Ca']= 75.0; p3['k_Ca']= 1e-4

        # Adaptive, repetitive firing profile
    elif profile=='adaptiveFiring':
        p3['A_NaT']= 2.5
        p3['A_KaD']= 20.0; p3['rate_Act_KaD']= 0.9;
        p3['A_KaSK']= 1.1
        p3['tau_Ca']= 1000.0; p3['k_Ca']= 1.5e-4
        
    # time-related parameters and initial conditions
    p3['timeMin']= 0.0
    p3['timeMax'] = 100.0
    p3['stepSize']= 1.0/40.0
    p3['timeSamples'] = np.arange(p3['timeMin'],p3["timeMax"],p3["stepSize"])
    p3['nSteps'] = len(p3['timeSamples'])
    p3['ic'] = np.array([0.005,-2.5,1e-4])
    #p3['rhs']=NaTKDSKCaL
    # stimulus parameters; start/stop times and amplitude
    p3['upStart']=200.0
    p3['upStop']=200.001
    p3['dnStart']=1000.0
    p3['dnStop']=1000.001
    p3['rampAmp']=100
    # functions to calculate Nernst potentials and normalize
    p3= calcReversalPotentials(p3)
    p3 = normalizeVolts(p3)
    return p3
```

```python
def pars_NeuronalBurster_PreSR(p,profile='condBursting',age='young'):
	p=pars_NeuronalBurster(p,profile,age)
	p['base_qSynapticRelease'] = 0.9
	p['tau_qSynapticRelease'] = 0.9
	p['c_Half_qSynapticRelease']= 0.1e-3
	p['gain_qSynapticRelease']=2.0
	p['q_qSynapticRelease'] = 0.9
	p['ic'] = np.array([0.005,-2.8,1e-7,0.9])
	return p

def pars_NeuronalBurster_PreSR_PostSC(p,profile='condBursting',age='young'):
	p=pars_NeuronalBurster_PreSR(p,profile,age)
	p['ic'] = np.array([0.005,-2.8,1e-7,0.9,0,0])
	return p

def printAmps_comparison(dicti):
	print('v_T = %g mV, Cm=%g pF, vTCm=%g mVpF, 1/(vTCm)=%g'%(dicti['v_T'],dicti['Cm'], dicti['vTCm'],1/dicti['vTCm']))
	print('&$a$ (pA) & $a/Cm$ (V/S) & A (1/ms) \\')
	print('\hline')
	for k,vv in dicti.items():
	    if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
	        print('%s & %g & %g & %g'%(k[2:],vv*dicti['vTCm'],vv/dicti['v_T'],vv))
	return

def printAmps_normvTCm(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'A_')==0) & (str.find(k,'A_Single_') !=0):
            print('%s = %g'%(k,vv))
    return

def printAmps_normCm(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'a_')==0) & (str.find(k,'a_Single_') !=0):
            print('%s/Cm = %g'%(k,vv/dicti['Cm']))
    return

def printAmps_pA(dicti):
    for k,vv in dicti.items():
        if (str.find(k,'a_')==0) & (str.find(k,'a_Single_') !=0):
            print('%s = %g'%(k,vv))
    return
```

```python
# -------------------------------------
# Plot the dynamics of v and c taking into account the change in v (color)
# -------------------------------------
def plot_Profile_PyrCA1(p3,s3,figName=''):
    figu3 = gr.figure(figsize=(27,11)); gr.ioff();
    hSpace, vSpace = 0.065,0.075
    height = 0.98*(1 - 3*vSpace)/2
    longW = 0.6
    shortW= 0.95*(1 - 2*hSpace - longW)
    ax_v_coords = [hSpace, 2*vSpace + height, longW, height]
    ax_dv_coords = [2*hSpace+longW, 2*vSpace + height, shortW, height]
    ax_i_coords = [hSpace, vSpace, longW, height]
    ax_idv_coords = [2*hSpace+longW, vSpace, shortW, height]
    ax_v = figu3.add_axes(ax_v_coords)
    bx_v = ax_v.twinx()
    ax_i = figu3.add_axes(ax_i_coords,sharex = ax_v)
    ax_dv = figu3.add_axes(ax_dv_coords,sharey = ax_v)
    ax_idv = figu3.add_axes(ax_idv_coords,sharex = ax_dv, sharey=ax_i)
    #figu3,ax_v = plot_Profile_PyrCA1(p3,s3, figu3, ax_v, cMap='viridis')
    s3['i_Tot']=  (s3['i_NaKa']+s3['i_NaT']+s3['i_KaD']+s3['i_KaSK']+s3['i_CaL'])
    bx_v.plot(s3['timeSamples'], s3['c'], 'orange', lw=4, alpha=0.5, label=r'$[Ca]_i$ ($\mu$M)')
    ax_v.plot(s3['timeSamples'], s3['v_Ca'], 'orange', lw=1, alpha=0.95, label=r'$v_{Ca}$ (mV)')
    ax_v.plot(s3['timeSamples'], s3['v'], 'k', lw=2, alpha=0.5, label=r'$v$ (mV)')
    #
    ax_i.plot(s3['timeSamples'], s3['i_NaT']/1000, 'b', alpha=0.5, label=r'$I_{NaT}$ ')
    ax_i.plot(s3['timeSamples'], s3['i_KaD']/1000, 'r', alpha=0.5, label=r'$I_{KaD}$ ')
    ax_i.plot(s3['timeSamples'], s3['i_KaSK']/1000, 'g', alpha=0.5, label=r'$I_{KaSK}$ ')
    ax_i.plot(s3['timeSamples'], s3['i_CaL']/1000, 'orange', lw=2, alpha=0.95, label=r'$I_{CaL}$ ')
    ax_i.plot(s3['timeSamples'], s3['i_Tot']/1000, 'k', lw=2, alpha=0.5, label=r'$I_{Tot}$ ')
    #
    ax_idv.plot(s3['\partial_t v'], s3['i_NaT']/1000, 'b', alpha=0.5, label=r'$(\partial_t v, I_{NaT})$')
    ax_idv.plot(s3['\partial_t v'], s3['i_KaD']/1000, 'r', alpha=0.5, label=r'$(\partial_t v,I_{KaD})$')
    ax_idv.plot(s3['\partial_t v'], s3['i_KaSK']/1000, 'g', alpha=0.5, label=r'$(\partial_t v,I_{KaSK})$')
    ax_idv.plot(s3['\partial_t v'], s3['i_CaL']/1000, 'orange', lw=2, alpha=0.95, label=r'$(\partial_t v,I_{CaL})$')
    #
    ax_dv.plot(s3['\partial_t v'], s3['v_Ca'], 'orange',lw=2, alpha=0.5, label=r'$(\partial_t v, v_{Ca})$ (mV)')
    ax_dv.plot(s3['\partial_t v'], s3['v'], 'k', lw=2, alpha=0.5, label=r'$(\partial_t v, v)$ (mV)')
    #ax_idv.plot(s3['i_Tot']/1000, s3['v'], 'k', alpha=0.5, label=r'$(I_{Tot},v)$')
    if s3['\partial_t v'].max()>100:
        ax_i.plot(s3['spikeTimes'], 5*np.ones(len(s3['spikeTimes'])),'k|', ms=10, label='Spike times')
        ax_v.plot(s3['spikeTimes'], 60*np.ones(len(s3['spikeTimes'])),'k|', ms=10,  label='Spike times')
        ax_v.plot(s3['spikeTimes'], s3['ifr'],'o', markeredgecolor='b',markerfacecolor='w', label='IFR (Hz)')
    ax_v.set_ylim(-90,200); bx_v.set_ylim(-1.5,1.2); bx_v.set_yticks(np.arange(0,1.1,0.3))
    ax_v.legend(loc='upper left',ncol=4);
    ax_v.set_yticks(np.arange(-80,160,20)); ax_v.set_ylabel('mV, Hz',rotation=0);
    ax_v.set_xlabel('msecs')
    bx_v.legend(loc='upper right'); bx_v.set_ylabel(r'$\mu$M',rotation=0)
    ax_i.legend(loc='upper right');
    ax_i.set_xlabel('msecs')
    ax_i.set_yticks(np.arange(-4,5)); ax_i.set_ylabel('nA',rotation=0)
    ax_dv.legend(loc='upper left'); 
    ax_idv.legend(loc='upper right'); ax_idv.set_xlabel('(V/s)')
    ax_i.set_ylabel('nA',rotation=0)
    figu3.subplots_adjust(top=0.96, bottom=0.075, left=0.055, right=0.98, hspace=0.1, wspace=0.1)
    gr.ion(); gr.draw(); gr.show()
    if len(figName)>0:
        figu3.savefig(figName)
    return figu3
```
